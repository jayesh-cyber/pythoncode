# Seaborn 

```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tip=sns.load_dataset('tips')
```

```python
tip.head()
```

```python
sns.countplot(x='sex',data=tip,palette='BuPu',hue='smoker')
plt.show()
```

```python
sns.boxplot(x='sex',y='total_bill',data=tip)
plt.show()
```

```python
sns.swarmplot(x='sex',y='total_bill',data=tip,hue='smoker')
plt.show()
```

```python
sns.violinplot(x='sex',y='total_bill',data=tip)
plt.show()
```

```python
plt.figure(facecolor='black')
sns.distplot(a=tip['total_bill'],hist_kws={'color':'red','edgecolor':'black'},
            kde_kws={'color':'g','label':'kde','lw':4},kde=True)
plt.savefig('C:/Users/chinki/Desktop/image.jpg',dpi=200)
```
